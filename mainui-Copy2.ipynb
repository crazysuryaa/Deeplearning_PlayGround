```python
class BotBubble:
    def __init__(self,master,root,message="",cond = "q"):
        self.master = master
        if cond == "q":
            self.frame = Frame(master,bg="light green")
            self.i = self.master.create_window(300,450,window=self.frame)
            Label(self.frame,text=datetime.now().strftime("%d-%m-%Y %X"),font=("Verdana bold", 7),bg="light green").grid(row=0,column=0,sticky="w",padx=5) #tarih saat        
            Label(self.frame, text=textwrap.fill(message, 25), font=("Verdana bold", 10),bg="light green").grid(row=1, column=0,sticky="w",padx=5,pady=3)
            root.update_idletasks()
            self.master.create_polygon(self.draw_triangle2(self.i), fill="light green", outline="light green")
        elif cond == "r" :
            self.frame = Frame(master,bg="#fed8b1")
            self.i = self.master.create_window(100,450,window=self.frame)
            Label(self.frame,text=datetime.now().strftime("%d-%m-%Y %X"),font=("Verdana bold", 7),bg="#fed8b1").grid(row=0,column=0,sticky="w",padx=5) #tarih saat        
            Label(self.frame, text=textwrap.fill(message, 25), font=("Verdana bold", 10),bg="#fed8b1").grid(row=1, column=0,sticky="w",padx=5,pady=3)
            root.update_idletasks()
            self.master.create_polygon(self.draw_triangle(self.i), fill="#fed8b1", outline="#fed8b1")


    def draw_triangle2(self,widget):
        x1, y1, x2, y2 = self.master.bbox(widget)
        return x1+90, y2 - 10, x1 + 180, y2 + 10, x1+90, y2


    def draw_triangle(self,widget):
        x1, y1, x2, y2 = self.master.bbox(widget)
        return x1, y2 - 10, x1 - 15, y2 + 10, x1, y2
```

```python
class PlotLosses(keras.callbacks.Callback):
    def on_train_begin(self, logs={}):
        self.i = 0
        self.x = []
        self.losses = []
        self.val_losses = []
        
        self.fig = plt.figure()
        
        self.logs = []

    def on_epoch_end(self, epoch, logs={}):
        
        self.logs.append(logs)
        self.x.append(self.i)
        self.losses.append(logs.get('loss'))
        self.val_losses.append(logs.get('val_loss'))
        self.i += 1
        
        clear_output(wait=True)
        plt.plot(self.x, self.losses, label="loss")
        plt.plot(self.x, self.val_losses, label="val_loss")
        plt.legend()
        plt.show();
        
plot_losses = PlotLosses()
```
