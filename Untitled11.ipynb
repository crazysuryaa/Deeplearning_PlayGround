```python
myclasses = [x[:-1] for x in open(dataset_train+'classes.txt','r').readlines()]
open(dataset_train+"labels.txt", 'w').close()
for folder in os.listdir(dataset_train):
#     print(folder)
    folderpath = dataset_train+folder
    if os.path.isdir(folderpath):
        for filename in os.listdir(folderpath):
            if  filename.endswith('.xml'):
                filepath = folderpath+"/"+filename
                bb = ""
                in_file = open(filepath)
                tree=ET.parse(in_file)
                root = tree.getroot()
                for i, obj in enumerate(root.iter('object')):
                    difficult = obj.find('difficult').text
                    cls = obj.find('name').text
                    print(cls)
                    if cls not in myclasses or int(difficult)==1:
                        continue
                    
                    cls_id = myclasses.index(cls)
                    xmlbox = obj.find('bndbox')
                    b = (int(xmlbox.find('xmin').text), int(xmlbox.find('ymin').text), int(xmlbox.find('xmax').text), int(xmlbox.find('ymax').text))
                    bb += (" " + ",".join([str(a) for a in b]) + ',' + str(cls_id))
#                 print(bb)
            list_file = open(dataset_train+"labels.txt", 'a')
            file_string = str(filepath)[:-4]+'.jpg'+bb+'\n'
            list_file.write(file_string)
            list_file.close()
```
