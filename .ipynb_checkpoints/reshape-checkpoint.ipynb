```python
    import cv2 
    import numpy as np 
    import matplotlib.pyplot as plt
```

```python
def createwidgets(path):

    img = cv2.imread(path)
    x,y = img.shape[0],img.shape[1]
#     img = cv2.resize(img , (70,70))
#     cv2.imwrite(path,img)
    img = cv2.resize(img , (y+10,x+10))
    img = cv2.rectangle(img,(0,0),(y+10,x+10),color = (0,255,0),thickness = 2 )
    # Add alpha layer with OpenCV
    
    bgra = cv2.cvtColor(img, cv2.COLOR_BGR2BGRA) 

    # Set alpha layer semi-transparent with Numpy indexing, B=0, G=1, R=2, A=3
    bgra[...,3] = 127

    bgra
    s = path[:-4]+"active"+path[-4:]
    cv2.imwrite(s,bgra)
    plt.imshow(bgra)
```

```python
createwidgets('/home/crazy/UI/widgets/inactive/add.png')
```

```python
def reshapebuttons(path,h,w,r=1,c=9):
    hr = int(h/r)
    wr = int(w/c)
    img = cv2.imread(path)
    x,y = img.shape[0],img.shape[1]
    img = cv2.resize(img,(hr,wr))
    cv2.imwrite(path,img)
```
