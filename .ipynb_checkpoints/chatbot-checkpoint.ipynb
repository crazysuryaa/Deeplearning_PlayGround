```python
from tkinter import *
from datetime import datetime
import random
import re
from tkinter import messagebox
from tkinter.font import Font
import textwrap
from random import choice                       
                                    
ask   = ["hi", "hello"]                         
hi    = ["hi", "hello"]
sq = ["whatsup", "kya kar rahe ho"]
rp = ["nothing u say", "kuch nhi ap btao"]
hr = ["who are you", "what is your name"]
rh = ["i am toko, created by surya"] 
emotion = ["i like you","i love you"]
express = ["i like you too"]                                         #you can add more conversation whatever you like 
error = ["sorry, i don't know", "what u said?","can't recognise" ]


root = Tk()
root.config(bg="lightblue")
root.geometry('410x600+400+100')

#ana ekran
canvas = Canvas(root, width=200, height=200,bg="white")
canvas.grid(row=0,column=0,columnspan=2)
canvas.place(x=10, y=10, width=390, height=530)
# while(1):     

bubbles = []

class BotBubble:
    def __init__(self,master,message="",cond = "q"):
        self.master = master
        if cond == "q":
            self.frame = Frame(master,bg="light green")
            self.i = self.master.create_window(300,490,window=self.frame)
            Label(self.frame,text=datetime.now().strftime("%d-%m-%Y %X"),font=("Helvetica", 7),bg="light green").grid(row=0,column=0,sticky="w",padx=5) #tarih saat        
            Label(self.frame, text=textwrap.fill(message, 25), font=("Helvetica", 9),bg="light green").grid(row=1, column=0,sticky="w",padx=5,pady=3)
            root.update_idletasks()
            self.master.create_polygon(self.draw_triangle2(self.i), fill="light green", outline="light green")
        elif cond == "r" :
            self.frame = Frame(master,bg="#fed8b1")
            self.i = self.master.create_window(100,490,window=self.frame)
            Label(self.frame,text=datetime.now().strftime("%d-%m-%Y %X"),font=("Helvetica", 7),bg="#fed8b1").grid(row=0,column=0,sticky="w",padx=5) #tarih saat        
            Label(self.frame, text=textwrap.fill(message, 25), font=("Helvetica", 9),bg="#fed8b1").grid(row=1, column=0,sticky="w",padx=5,pady=3)
            root.update_idletasks()
            self.master.create_polygon(self.draw_triangle(self.i), fill="#fed8b1", outline="#fed8b1")

    
    def draw_triangle2(self,widget):
        x1, y1, x2, y2 = self.master.bbox(widget)
        return x1+90, y2 - 10, x1 + 180, y2 + 10, x1+90, y2


    def draw_triangle(self,widget):
        x1, y1, x2, y2 = self.master.bbox(widget)
        return x1, y2 - 10, x1 - 15, y2 + 10, x1, y2

def send_message():
    if bubbles:
        canvas.move(ALL, 0, -80) #balonlar arasındaki mesafe
    question =entry.get()
#     query = speech_to_text()
    a = BotBubble(canvas,question,"q")
    bubbles.append(a)
    getanswer(question)
#     reply_message(question)

def reply_message(question):
    if bubbles:
        canvas.move(ALL, 0, -80) #balonlar arasındaki mesafe
        reply = entry.get() 
#         if question in ask:                      
#              reply = choice(hi)
#         elif question in sq:
#              reply = choice(rp)
#         elif question in hr:
#              reply = choice(rh)
#         elif question in emotion:
#              reply = choice(express)
#         else:                                
#              reply = choice(error) 
        a = BotBubble(canvas,reply,"r")
        bubbles.append(a)
        

def speech_to_text():
    required=-1
    for index, name in enumerate(sr.Microphone.list_microphone_names()):
        if "pulse" in name:
            required= index
    r = sr.Recognizer()
    with sr.Microphone(device_index=required) as source:
        r.adjust_for_ambient_noise(source)
        print("Say something!")
        audio = r.listen(source, phrase_time_limit=4)
    try:
        query = r.recognize_google(audio)
        query = query.lower()
        print("You said: " + query)
        return query
    except sr.UnknownValueError:
        print("Google Speech Recognition could not understand audio")
    except sr.RequestError as e:
        print("Could not request results from Google Speech Recognition service; {0}".format(e))
    

def speak(audio):
    print('TOKO: ' + audio)
    entry.insert(0,audio)
    reply_message(audio)
    entry.delete(0,END)
    engine.say(audio)
    engine.runAndWait()

#mesaj yazma alanı
entry = Entry(root,width=26, font=("Helvetica", 10))
entry.place(x=10, y=550, width=290, height=40)


    

#buton
buton = Button(root, width=10, height=2, relief='raised',state='active',command=send_message)
buton.config(text='ASK', bg='lightblue', font='Verdana 8 bold')
buton.place(x=310, y=550)


root.mainloop()
```
